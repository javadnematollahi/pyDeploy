```python
import requests
import dotenv
import os
```

## Lord of Ring

## Icon finder

```python
url = "https://api.iconfinder.com/v4/icons/search"
dotenv.load_dotenv()
icon_key = os.getenv("icon_finder")
headers = {
    "accept": "application/json",
    "Authorization": f"Bearer {icon_key}"
}
payload = {
    "query":"arrow",
    "count":"10"
}
response = requests.get(url, headers=headers, params=payload)

print(response.status_code)
```

```python
response.json()
```

## d-id.com
